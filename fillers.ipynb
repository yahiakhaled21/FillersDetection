```python
import whisper_timestamped as whisper
```

```python
audio = whisper.load_audio("/kaggle/input/test-for-fillers/test_for_fillers.wav")
model = whisper.load_model("large")
result = whisper.transcribe(model, audio, language="en", detect_disfluencies=True)
print(result)
```

```python
filler_counter = 0
filler_words=[]
for text in result["segments"]:
    for word in text["words"]:
        if "[*]" in word["text"]:
            filler_words.append([word["start"], word["end"]])
            filler_counter = filler_counter + 1
```
